```python
history_length = []

with open("/home/duccd/Downloads/MINDsmall_train/behaviors.tsv") as file:
    for line in file:
        l = line.split('\t')
        history = l[3].split()
        history_length.append(len(history)) 
```

```python
from collections import Counter

w = sorted(Counter(history_length).items())
```

```python
w = Counter(history_length)
w
```

```python
for i in range(151, 1000):
    try:
        w.pop(i)
    except:
        print(i)
```

```python
import matplotlib.pyplot as plt

plt.bar(w.keys(), w.values())
plt.xlabel('Độ dài history')
plt.ylabel('Số lượng impression cùng độ dài')
plt.show()
```

```python
news_length = []

with open("/home/duccd/Downloads/MINDsmall_train/news.tsv") as file:
    for line in file:
        l = line.split('\t')
        news_length.append(len(l[3].split()))

from collections import Counter

n = Counter(news_length)
```

```python
import matplotlib.pyplot as plt

plt.bar(n.keys(), n.values(), color='green')
plt.xlabel('Độ dài news_title')
plt.ylabel('Số lượng bài báo cùng độ dài title')
plt.show()
```
